```julia
∂ₜK + ∂ₓK = -1/√2 ̄HA
∂ₜR - 1/3*∂ₓR = -2√2/3 ̄HA
∂ₜQ + ̄Q/√2∂ₓK - Q/6√2∂ₓR = (̄Q/6 - 1)̄HA
∂ₜA = ΓQAₛ
```

```julia
function smoothing(field, num_k, alpha, Nx)
    # for a signal (field) of size Nx
    # retains only the first num_k modes 
    # all non-zero modes are damped by alpha (ie variations around the mean are reduced)
    # returns the smoothed field and its fourier coefficients in order -1 -->-kM, kM -->0  
    idx0 = Int(floor(Nx/2+1));
    
    rft = fftshift(fft(field));
    rft[1:idx0-num_k-1] .= 0.0;
    rft[idx0+num_k+1:end] .= 0.0;
    rft[idx0+1:end] .= alpha*rft[idx0+1:end];
    rft[1:idx0-1] .= alpha*rft[1:idx0-1];
    Field_smoothed = ifft(ifftshift(rft));
    
    Fhat = reverse(ifftshift(rft)); # order: -1 -->-kM, kM -->0  
    
    return Field_smoothed, Fhat
end
```

```julia
# smooth background for the computation of initial conditions
num_k_IC = 1;
alpha_IC = 0.0;
Aₛ_IC, Shat_IC = smoothing(mean(Aₛ[:,:], dims = 1)[1,:]/γ, num_k_IC, alpha_IC, Nx); 
```

manual reconstruction, important to undestand what is going on and have a rererence for coefficients (division by Nx etc)

```julia
κ=2*π*im*fftshift(fftfreq(Nx,Nx))/L;

xx = collect(0:Nx-1)*dx;
As_reconstructed_manually = zeros(Nx)*im;
coefs = fftshift(reverse(Shat_IC));
for i in 1:Nx
    As_reconstructed_manually += coefs[i]/Nx*exp.(κ[i]*xx)*γ;
end
As_reconstructed_manually = real(As_reconstructed_manually);


As_reconstructed = ifft(reverse(Shat_IC))*γ;#

plot(longitudes, Hbar*real(As_reconstructed), color = "magenta")
plot(longitudes, Hbar*real(As_reconstructed_manually), color = "blue")
plot(longitudes, Hbar*real(Aₛ[1,:]), color = "grey")
```

define matrices for the eigenproblem

```julia
AAA = zeros(4*size(κ)[1], 4*size(κ)[1]).*im;
SSN = zeros(4, 4*(Nx-1)).*im;
SSP = zeros(4, 4*(Nx-1)).*im;

# separate negative and positive values in the frequencies ordered as -1 -->-kM, kM -->0
κ2 = reverse(ifftshift(fftshift(fftfreq(Nx,Nx))));
idx0 = findall(x -> x == -Int(floor(Nx/2)), κ2);
for j in 1:Nx-1
    if j <=idx0[1]
        SSN[4, 3+4*(j-1)] = im*Γ*Shat_IC[j]/Nx
    elseif j > idx0[1]
        SSP[4, 3+4*(j-1)] = im*Γ*Shat_IC[j]/Nx
    end
end


for (i, κi) in enumerate(κ[1:end])
    #### diagonal
    tmp = [κi           0.0           0.0                1/√2*Hbar;
           0.0         -1/3.0*κi      0.0                2*√2/3.0*Hbar;
           Qbar/√2*κi  -Qbar/6/√2*κi  0.0               -(Qbar/6.0- 1.0)*Hbar;
        0.0          0.0          -Γ*Shat_IC[end]/Nx         0.0];
    tmp = -im.*tmp; # so that the system becomes AU = ωU
    
    ### lower triangle
    #println("filling rows : ", i+3*(i-1), " to ", i+3*(i-1)+3 )
    if i>1
        #println("bloc: ", 1, " to ", i+3*(i-1)-1, " with wavenumbers ", real(SS[:, end - 4*(i-1)+1 : end][4,3]), " to ", real(SS[:, end - 4*(i-1)+1 : end][4,end-1]))
        AAA[i+3*(i-1):i+3*(i-1)+3, 1:i+3*(i-1)-1] = SSP[:, end - 4*(i-1)+1 : end]
    end
    
    ### upper triangle
    #println("bloc: ",  i+3*(i-1), " to ", i+3*(i-1)+3)
    AAA[i+3*(i-1):i+3*(i-1)+3, i+3*(i-1):i+3*(i-1)+3] = tmp;    
    if i<Nx
        #println("bloc: ", i+3*(i-1)+4, " to ", 4*Nx, " with wavenumbers ", real(SS[:,1:4*Nx - (i+3*(i-1)+3)][4,3]), " to ", real(SS[:,1:4*Nx - (i+3*(i-1)+3)][4,end-1]))
        AAA[i+3*(i-1):i+3*(i-1)+3, i+3*(i-1)+4:4*Nx] = SSN[:,1: 4*Nx - (i+3*(i-1)+3)]   
    end
end

V = eigvecs(AAA);
d = eigvals(AAA);   

```

```julia

#### ccompute "mean" k for each of the solution vector, we'll take the one closest to 2 as initial condition for the MJO mode

# factors
factors = zeros(Nx,4*Nx);
for j in 1:4*Nx
    tot = 0.0;
    for i in 1:Nx        
        tot = tot + norm(V[i+3*(i-1):i+3*(i-1)+3,j]);
        factors[i, j] =  norm(V[i+3*(i-1):i+3*(i-1)+3,j]);
    end
    factors[:,j] = factors[:,j]./tot;
end

kk = zeros(4*Nx)
for j in 1:4*Nx
    ks = 0.0;
    for i in 1:Nx        
        ks = ks + (imag(κ[i])/(2.0*pi/L))*factors[i,j];
    end
    kk[j]=ks;
end


cpd = (d)*(24/8.333)/(2*pi);
vel = real(cpd .*40000000/60/60/24 ./kk);

#----------------------------------------------------------
# select the 4 modes for each k (Kelvin, dry and moist Rossby, MJO)
#-----------------------------------
kk1 = kk[1:Nx]; 
kk2 = kk[Nx+1:2*Nx]; 
kk3 = kk[2*Nx+1:3*Nx]; # from indices Nx/2 to Nx --> MJO
kk4 = kk[3*Nx+1:4*Nx];

ω1 = cpd[1:Nx];
ω2 = cpd[Nx+1:2*Nx];
ω3 = cpd[2*Nx+1:3*Nx]; # from indices Nx/2 to Nx --> MJO
ω4 = cpd[3*Nx+1:4*Nx];

vel1 = vel[1:Nx];
vel2 = vel[Nx+1:2*Nx];
vel3 = vel[2*Nx+1:3*Nx]; # from indices Nx/2 to Nx --> MJO
vel4 = vel[3*Nx+1:4*Nx];

V1 = (V[:,1:Nx]);
V2 = (V[:,Nx+1:2*Nx]);
V3 = (V[:,2*Nx+1:3*Nx]); # from indices Nx/2 to Nx --> MJO
V4 = (V[:,3*Nx+1:4*Nx]);

idxMJO2 = argmin(abs.(kk3.-2));
print("closest wavenumber to 2 is ", kk3[idxMJO2]);
```

```julia
initampl = 0.05; 

K0 = zeros(Nx, 1).*im;
R0 = zeros(Nx, 1).*im;
Q0 = zeros(Nx, 1).*im;
Z0 = zeros(Nx, 1).*im;
A0 = zeros(Nx, 1).*im;

xx = collect(0:Nx-1)*dx;
yy = range(-3,3, length = 40);
phi0 = γ.*exp.(-yy.^2 ./2.0);
phi2 = γ.*1.0/√2 .* (2.0.*yy.^2 .-1.0) .*exp.(-yy.^2 ./2.0);

contributionsK=zeros(Nx);
contributionsR=zeros(Nx);
contributionsQ=zeros(Nx);
contributionsA=zeros(Nx);

for i in 1:Nx
    vec = V3[i+3*(i-1):i+3*(i-1)+3, idxMJO2];
    K0[:,1] += initampl*vec[1]*exp.(κ[i]*xx);
    R0[:,1] += initampl*vec[2]*exp.(κ[i]*xx);
    Q0[:,1] += initampl*vec[3]*exp.(κ[i]*xx);
    A0[:,1] += initampl*vec[4]*exp.(κ[i]*xx); 
    contributionsK[i] = norm(vec[1]); 
    contributionsR[i] = norm(vec[2]); 
    contributionsQ[i] = norm(vec[3]); 
    contributionsA[i] = norm(vec[4]); 
    
end    

K0_ref =K0;
R0_ref =R0;
Q0_ref =Q0;
A0_ref =A0;


K0 = real(K0);
R0 = real(R0);
Q0 = real(Q0);
A0 = real(A0);

a = A0' .* phi0; #.*15.0*24.0/8.0*Hbar
q = Q0' .* phi0;
theta = -1/√2 * (K0' .+ R0'./2.0).*phi0 .- R0'./4.0.*phi2 ;
p = -theta;
```
